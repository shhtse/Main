# Human language -> ER-diagram --> SQL
A - relation = B
> $\exists$ a $\in$ A s.t. a $\notin$ Range(B)

> $\forall$ b $\in$ B $\exists$ a $\in$ A s.t. A(a) = b

A n relation m B
> $\exists$ n number of b$\in$B s.t. A(a) = b

> $\exists$ m number of a$\in$A s.t. B(b) = a

A is weak entity of B
> A $\subset$ B(b)

# Create

CREATE TABLE name1(
    Col1 INT,
    Col2 VARCHAR(20),
    PRIMARY KEY(Col1)
    FOREIGN KEY (col1) REFERENCE foreign(fcol1) ON DELETE SET NULL
);
> Create a table
> Columns contain non-unique element cannot be the primary key, but composite of them can be
> FOREIGN KEY = restrict the elements of this column must belong to the col of other table
> ON DELETE SET NULL = if the original datum has been deleted, preserve the row set the deleted datum as null
> ON DELETE CASCADE = if the original datum has been deleted, delete the corresponding row as well

CREATE TABLE name(
